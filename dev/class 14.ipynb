# Integration

average of integrals == midpoints == trapezoid rule

integral(f(x))                                                                       \
    == lim as n->infinity sum(f(xn), n=0, n-1)\*h                                    \
    == lim as n->infinity sum(f(xn, n=1, n))\*h                                      \
    ~~ lim as n->infinity ((sum(f(xn), n=0, n-1) + sum(f(xn, n=1, n))) / 2) \*h      \
    ~~ h/2\[f(a)+f(b)+2f(x)+2f(x2)+2f(x3)\]                                          \
trapezoid rule                                                                       \
    ~~ h/2\[f(a)+f(b)\] + sum(f(xn, n=1, n-1)) - h  

```python
import numpy as np
import pandas as pn
import matplotlib.pyplot as plt
import math
```

```python
def myFunc(x):
    return x**3

def sinlis(x):
    l = []
    for i in x:
        l = np.append(l, math.sin(i))
    return l

a,b,steps = 1,2,10

x = np.linspace(a, b, steps+1)
```

```python
def integrate(func, a, b, steps):
    x = np.linspace(a, b, steps+1)
    y = func(x)
    
    h = (b-a)/steps

    s_terms = (func(a)+func(b))*h/2
    
    #for i in x[1:-1]:
    #    s = s + func(i)*h
    
    #s_total = s_terms + np.sum(func(x[1:-1])*h)
    
    s_total = s_terms + np.sum(y[1:steps]*h)
    
    return s_total
```

```python
def integrateGraph(func, a, b, steps):
    x = np.linspace(a, b, steps+1)
    # y = func(x)
    # h = (b-a)/steps
    # s_terms = (func(a)+func(b))*h/2

    s = []
    for i in x:
        s.append(integrate(func, i, i+1, steps))
        
    
    return s
```

```python
integrateGraph(myFunc, a, b, steps)
```

```python
fig0, ax0 = plt.subplots()

ax0.plot(x,myFunc(x), ".-", label="f(x)")
ax0.plot(x,integrateGraph(myFunc, a, b, steps+1)[0:-1], ".-", label="S(f(x))")
ax0.legend()
```

```python
fig1, ax1 = plt.subplots()

ax1.plot(x,myFunc(x), ".", label="f(x)")
ax1.plot(x,sinlis(x), ".", label="S")
ax1.legend()
```
