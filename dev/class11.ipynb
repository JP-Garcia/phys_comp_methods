```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def forwardDiff(func, x, h):
    return((func(x+h)-func(x))/h)
def backwardDiff(func, x, h):
    return((func(x)-func(x-h))/h)
def midpointDiff(func, x, h):
    return((func(x+h)-func(x-h))/2/h)
def doubleDiff(func, x, h):
    return((func(x+h)-2*func(x)+func(x-h))/h**2)
```

```python
def myFunc(x):
    fx = (1/3)*x**3
    return fx
```

```python
myFunc(1)
```

```python
forwardDiff(myFunc, 1, .01)
```

```python
backwardDiff(myFunc, 1, .01)
```

```python
midpointDiff(myFunc, 1, .01)
```

```python
doubleDiff(myFunc, 1, .01)
```

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot()
x = []
y = []

j = .1
for i in range(21):
    print(midpointDiff(myFunc, 1, j))
    j = j/10
    x.append(i)
    y.append(midpointDiff(myFunc, 1, j))
    
ax3.plot(x, y, "o-")
```

# Plot function and it's derivative

```python
def derivative(func, v0, vf):
    x = np.linspace(v0, vf)
    y = []
    dy = []
    for i in x:
        y.append(func(i))
        dy.append(midpointDiff(func, i, .001))
    return x,y,dy
```

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot()

x,y,dy = derivative(myFunc, 0, 5)
ax3.plot(x, y, ".-", )
ax3.plot(x, dy)
```
