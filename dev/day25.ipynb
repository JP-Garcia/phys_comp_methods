# Runga-Kutta 2nd derivative ODE solver

d^2y / d^2t = -g

d^2y / d^2t = dv / dt

dv / dt = -g

### Change Runga-Kutta: change x to y, y to v, a and b to t0 and tf

```python
def rungaFourth(f, y0=0, v0=0, t0=0.0, tf=10.0, x=0.0, N=1000):
    dt = (tf-t0)/N
    
    r = np.array([y0,v0], float) #init conditions

    tpoints = np.arange(t0,tf,dt)
    ypoints = []
    vpoints = []

    for t in tpoints:
        ypoints.append(r[0])
        vpoints.append(r[1])
        k1 = dt*f(r,t)
        k2 = dt*f(r+0.5*k1, t+0.5*dt)
        k3 = dt*f(r+0.5*k2, t+0.5*dt)
        k4 = dt*f(r+0.5*k3, t+0.5*dt)
        r = r + (k1+2*k2+2*k3+k4)/6
    
    return tpoints, ypoints, vpoints
```

```python
t,y,v = rungaFourth(f, y0=100, tf=4.5)
fig0, ax0 = plt.subplots()
ax0.plot(t,y)
```
