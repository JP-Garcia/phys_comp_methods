# Eulers methood

## first order


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#derivative of function we want
def f(x,t): return(-x**3 + np.sin(t))

# set boundry conditions

a = 0.0 # start
b = 3.0 # end
N = 1000 # num of poins between a and b
dt = (b-a)/N

x = 0.0 # init


tpoints = np.arange(a,b,dt)
xpoints = []

for t in tpoints:
    xpoints.append(x)
    x = x + dt * f(x,t) # next x = x + dt * derivative
```

```python
fig0, ax0 = plt.subplots()

ax0.plot(tpoints, xpoints)
ax0.set_xlabel('time')
ax0.set_ylabel('position')
```

```python
def eulerFirst(f, a, b):
    N = 1000 # num of poins between a and b
    dt = (b-a)/N
    x = 0.0 # init

    tpoints = np.arange(a,b,dt)
    xpoints = []

    for t in tpoints:
        xpoints.append(x)
        x = x + dt * f(x,t) # next x = x + dt * derivative
    
    return tpoints, xpoints
```

```python
def graph(x, y, xname, yname):
    fig0, ax0 = plt.subplots()

    ax0.plot(x, y)
    ax0.set_xlabel(xname)
    ax0.set_ylabel(yname)
```

```python
t,x = eulerFirst(f, 0, 3)
```

```python
graph(t, x, "time", "position")
```

## second order eqn

```python
def f2(x, v, t): return(-9.8)
```

```python
def eulerSecond(f, a, b):
    N = 1000 # num of poins between a and b
    dt = (b-a)/N
    x = 10.0 # init
    v = 0.0

    tpoints = np.arange(a,b,dt)
    xpoints = []
    vpoints = []

    for t in tpoints:
        xpoints.append(x)
        vpoints.append(v)
        v = v + dt * f(x,v,t) # next x = x + dt * derivative
        x = x +dt * v
        # if (x <= -0.1):
        #     break
    
    return tpoints[0:len(xpoints)], xpoints
```

```python
def graph(x, y, xname, yname):
    fig0, ax0 = plt.subplots()

    ax0.plot(x, y)
    ax0.set_xlabel(xname)
    ax0.set_ylabel(yname)
```

```python
t,x = eulerSecond(f2, 0, 1.41)
```

```python
graph(t, x, "time", "position")
```
