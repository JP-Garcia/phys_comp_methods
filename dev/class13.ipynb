```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Create data
x = np.linspace(0,3*math.pi, 100)
y = np.cos(x)
```

```python
ydiff = np.gradient(y) #difference between y[n+1] and y[n]

dy = np.gradient(y, .05)
```

```python
#   Gradient Function:
#   (y[n+1] - y[n]) / (x[n+1] - x[n])
#
#   x     y    dy
#   0     3    NaN (Not a Number, or random algorithm??)
#   .05   5    2/.05
#   .1    7    2/.05
```

```python
# new way of graphing
fig0, ax0 = plt.subplots()

ax0.plot(x,y, "o", label=r"cos($\theta)")
ax0.plot([0,3*math.pi], [0,0])
ax0.plot(x, dy, "^")
ax0.legend()

# ax0.legend(["cos(x)", "x=0", "dx/dy"])
```
