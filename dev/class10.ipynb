## Least Squares - Milikan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mx, my = np.loadtxt('../data/millikan.txt', unpack=True)
print(mx)
print(my)
```

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot()

ax3.plot(mx, my, "o")
# ax3.plot(mx, my)
# ax3.plot([0,1e15],[0,1], '-g')
```

```python
ex = (1/len(mx)) * sum(mx)
ey = (1/len(my)) * sum(my)
exx = (1/len(mx)) * sum(mx**2)
exy = (1/len(my)) * sum(mx*my)

m = (exy - ex*ey)/(exx - ex**2)
c = (exx*ey - ex*exy)/(exx - ex**2)

print(m)
print(c)
```

```python
fig4 = plt.figure()
ax4 = fig4.add_subplot()

fit_y=m*mx+c

ax4.plot(mx, my, "o")
ax4.plot(mx, fit_y, ".-")
```
