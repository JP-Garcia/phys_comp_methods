```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/170330-00.mpt',encoding='windows-1252', delimiter='\t', header=55)
```

```python
data
```

```python
data.columns
```

```python
fig0 = plt.figure()
ax0 = fig0.add_subplot()

ax0.semilogx(data['freq/Hz'], data['Re(Z)/Ohm'])

ax0.set_xlabel('freq/Hz')
ax0.set_ylabel('Re(Z)/Ohm')
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot()

ax1.semilogx(data['freq/Hz'], data['-Im(Z)/Ohm'], "v-g")

ax1.set_xlabel('freq/Hz')
ax1.set_ylabel('-Im(Z)/Ohm')
```
